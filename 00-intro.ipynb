# Delfos M1 — Fundamentos, Arquitectura y Gobernanza
### Nequi · Plataforma de datos

**Delfos** es la plataforma de datos de Nequi: un conjunto de principios, herramientas
y contratos que permite a cada equipo publicar, consumir y gobernar datos con autonomia,
sin sacrificar la consistencia regulatoria que exige la Superintendencia Financiera de Colombia (SFC).

## Prerrequisitos

- Acceso al workspace de Databricks con Unity Catalog habilitado.
- Bucket S3 aprovisionado (ver carpeta `infraestructura/` en el repositorio).
- Credenciales AWS configuradas como Databricks Secrets en el scope `nequi`.
- Airflow corriendo en la instancia EC2 del ambiente de taller (ver `infraestructura/dags/`).

---

## Como empezar

1. Ingresa el nombre de tu bucket S3 en el widget **s3_bucket** (parte superior del notebook).
2. Abre **01-data-mesh-delfos** en una pestana nueva del workspace.
3. Haz clic en **Run All** — el notebook `_resource/00-setup` se ejecuta automaticamente en la primera celda.
4. Sigue el orden de los modulos: **01 → 02 → 03 → 04**.

> **Nota:** cada modulo comienza con `%run ./_resource/00-setup`, que configura el catalogo,
> las rutas S3 y las funciones compartidas. No es necesario ejecutar el setup por separado.
